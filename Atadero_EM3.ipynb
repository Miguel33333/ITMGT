```python
# write code here

#define functions to check carrier type
def carrier_type(phonenumber):
    if phonenumber == "0918":
        return "smart"
    else:
        return "globe"
```

```python
# start stop and step variables 
start, stop, step = 0, 4, 1
  
# slicing first 4 elements 
final_recweek["phonenumber"]= final_recweek["phonenumber"].str.slice(start, stop, step) 
  
#first 4 numbers of each phonenumber
final_recweek
```

```python
#Vectorize
final_recweek['phonenumber'].apply(carrier_type)

#add column for carrier
final_recweek['carrier'] = final_recweek['phonenumber'].apply(carrier_type)

#dump
final_recweek
```
